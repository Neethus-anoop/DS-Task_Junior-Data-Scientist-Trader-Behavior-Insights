# Bitcoin Sentiment vs Trader Performance Analysis


# Analyze how market sentiment (Fear/Greed) affects trader performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Datasets (Update file paths)

```python
trades = pd.read_csv('historical_data.csv')
sentiment = pd.read_csv('fear_greed.csv')
trades.head(), sentiment.head()
```

## Data Cleaning

```python
trades['time'] = pd.to_datetime(trades['time'])
sentiment['Date'] = pd.to_datetime(sentiment['Date'])
trades['date'] = trades['time'].dt.date
sentiment['date'] = sentiment['Date'].dt.date
```

## Merge Data

```python
df = pd.merge(trades, sentiment, on='date', how='left')
df.head()
```

## Analysis

```python
# Avg PnL
df.groupby('Classification')['closedPnL'].mean()
```

```python
# Total PnL
df.groupby('Classification')['closedPnL'].sum()
```

```python
# Win Rate
df['win'] = df['closedPnL'] > 0
df.groupby('Classification')['win'].mean()
```

```python
# Leverage
df.groupby('Classification')['leverage'].mean()
```

## Visualization

```python
df.groupby('Classification')['closedPnL'].mean().plot(kind='bar')
plt.title('Average PnL by Sentiment')
plt.show()
```
